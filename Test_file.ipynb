```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv')
print(df.head())
```

```python
#printing columns
print(df.columns)
```

```python
# printing specific columns
print(df[['Name','Attack','Type 1' ]])
```

```python
#printing rows
print(df.iloc[0:4])
```

```python
# Read a specific location (R,C)
print(df.iloc[2,1])
```

```python
for index, row in df.iterrows():
    print(index, row['Type 1'])
```

```python
#Accessing rows based on a particular column
df.loc[df['Type 1'] == 'Grass']
```

```python
#sorting data
df.sort_values('Name', ascending=False)
df.sort_values(['Type 1', 'HP'], ascending=[1,0]) # sorted with Type 1 ascending and HP descending
```

```python
#adding a new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head()
```

```python
#dropping columns
df = df.drop(columns=['Total'])
```

```python
df.head()
```

```python
# Faster way to add a new column
df['Total'] = df.iloc[:,4:10].sum(axis=1) # axis = 1 adds horizontally axis = 0 adds vertically
df.head()
```

```python
#cols = list(df.columns)  gets all the columns in a list
#df.to_csv('name you give the file')  modifying / creating a file
#filtering Data
df.loc[df['Type 1'] == 'Grass']
```

```python
#filtering with one or more conditions
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] >70)] # and in pands is & or in |
```

```python
new_df =df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] >70)]
```

```python
new_df.reset_index()
```

```python
new_df.reset_index(drop=True, inplace=True) #drop old index
```

```python
#filtering with a specific sub-string contained in a string
df.loc[df['Name'].str.contains('Mega')]
# the opposite
df.loc[~df['Name'].str.contains('Mega')]
```

```python
import re  #regular expressions
df.loc[df['Type 1'].str.contains('fire| glass', flags=re.I, regex=True)]
```
